# Problem Set 4: Predict the Top Complaint Type

```python
# Import your libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as date

# Import all the ML libraries likely to be used 
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures 
```

```python
# load 311 data into a dataframe
data = r"311_Service_Requests.csv"
df = pd.read_csv(data)
# Remembering to perform this step from problem set 1
df['Complaint Type'] = np.where(df['Complaint Type']=='HEATING','HEAT/HOT WATER',df['Complaint Type'])
```

So far, we've identified the most common complaint type, isolated regions with the most common complaints, and attempted to find a correlation between building characteristics and complaint types. Now, we want to <b>build a predictive model for future prediction of the possibility of the previously identified top complaint type</b>

We do this once more by studying the 311 dataset. Using the best ML model, we'll predict the number of future complaints.

We'll start by just using a dataframe with heating complaints (the tope identified complaint from problem set 1).

```python
df_heat = df[df["Complaint Type"] == "HEAT/HOT WATER"]

# Convert into datetime
df_heat['Created Date'] = pd.to_datetime(df_heat['Created Date'])
```

### Building a predictive model by year

```python
# Now group all complaints by year
df_heat_year = df_heat.groupby(df_heat["Created Date"].dt.year)
df_heat_year.count()
```

```python
df_heat_year_=df_heat_year.count()
```

```python
# Purposely going to remove complaints from 2020 because it's obviously incomplete
plt.scatter(df_heat_year_.index[:-1],df_heat_year_['Complaint Type'][:-1])
plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of Heating Complaints') # add y-label to the plot
plt.title('Number of Heating Complaints by Year') # add title to the plot
plt.show()
```

```python
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(8),
                           LinearRegression())

#Let's define X and Y
X = df_heat_year_.index[:-1]
y = df_heat_year_["Complaint Type"][:-1]
poly_model.fit(X[:, np.newaxis], y)
yfit = poly_model.predict(X[:, np.newaxis])

plt.scatter(X, y,color='blue')
plt.plot(X, yfit,color='red');
plt.title('Regression') 
plt.xlabel('YEARS') 
plt.ylabel('Number of Heating Complaints') 
plt.show()
```

Now, we can use the ten 8 year trend to predict 2019 and 2020 heating complaints (let's pretend that COVID-19 didn't happen)

```python
# 2020 prediction
poly_model.predict(2020)
```

```python
# 2021 prediction
poly_model.predict(2021)
```

### Building a predictive model by month

```python
# Now group all complaints by year
df_heat_month = df_heat.groupby(df_heat["Created Date"].dt.month)
df_heat_month_=df_heat_month.count()
df_heat_month.count()
```

```python
#Let's define A and B
A = df_heat_month.count().index[:]
B = df_heat_month_["Complaint Type"]
```

```python
# Let's see how the trend looks like month to month before we select a predictor
plt.scatter(df_heat_month_.index,df_heat_month_['Complaint Type'])
plt.xlabel('Month') # add to x-label to the plot
plt.ylabel('Number of Heating Complaints') # add y-label to the plot
plt.title('Number of Heating Complaints by Year') # add title to the plot
plt.show()
```

Obviously, the number of heating complaints will always peak around the winter remain low in the summer.

```python
# Import a linear regression model
lin2 = LinearRegression()
#A = A[:, np.newaxis]
#A = [A.tolist()]
A = np.asarray(A)
A = A.reshape(-1,1)
# Import polynomial regression model
poly2 = PolynomialFeatures(degree = 2) 
A_poly = poly2.fit_transform(A) 
 
lin2.fit(A_poly, B)
```

```python
# Visualising the Polynomial Regression results 
plt.scatter(A, B, color = 'blue') 
  
plt.plot(A, lin2.predict(A_poly), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Month') 
plt.ylabel('Number of Heating Complaints') 
  
plt.show()
```

```python
# Predicting the average heating complaints in June across NYC
lin2.predict(poly2.fit_transform([[6]]))
```
