# Problem Set 3: What Is the Relationship between Housing Characteristics and Complaints? 

```python
# Import your libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load 311 data into a dataframe
data = r"311_Service_Requests.csv"
df = pd.read_csv(data)
# Remembering to perform this step from problem set 1
df['Complaint Type'] = np.where(df['Complaint Type']=='HEATING','HEAT/HOT WATER',df['Complaint Type'])
# Create a dataframe just for complaints from the Bronx
df_bronx = df[df["Borough"] == "BRONX"]
# Create a separate dataframe just for heating complaints from the Bronx
df_bronx_heating = df_bronx[df_bronx["Complaint Type"] == "HEAT/HOT WATER"]
```

With the most common complaint identified, the goal of this exercise is to determine <b> if the Complaint Type that you identified in response to Question 1 has an obvious relationship with any particular characteristic of the type of house or its designed construction?</b>

We do this once more by studying the 311 dataset as well as the PLUTO dataset which provides more context to houses in each borough. For this exercise, we use the most affected borough identified in Question 2.

```python
# Load pluto data for the Bronx in a dataframe
data_bronx = r"pluto_data\BX_18v1.csv"
bronx_pl_df = pd.read_csv(data_bronx)
bronx_pl_df.head()
```

The pluto dataset simply provides us with more information than we need. Therefore, we only select those columns pertinent to the current operation.

```python
bronx_df_ = bronx_pl_df[["Address","BldgArea", "BldgDepth", "BuiltFAR", "CommFAR", "Lot", "LotArea", "LotDepth", "NumBldgs", "NumFloors", 
             "OfficeArea", "ResArea", "ResidFAR", "RetailArea", "YearBuilt","YearAlter1","YearAlter2", "ZipCode", "YCoord", "XCoord"]]
bronx_df_.info()
```

```python
# While we're here, we're also going to drop duplicate addresses
bronx_df_.drop_duplicates(subset="Address", keep=False, inplace=True)
```

The other 311 database importantly gives us the number of heating complaints directly corresponding to their respective addresses. We can respectively summarize that database into a new one identifying the number of complaints to each address. In a city like New York with great populations living in buildings as opposed to single units, this is necessary.

```python
# Before moving forward, ensure that the address field in the dataset is filled in everywhere
df_bronx_heating["Incident Address"].isnull().sum()
# If not, drop all cases with no tied address
df_bronx_heating.dropna(subset = ["Incident Address"], axis = 0, inplace=True)

# Now, we'll grab the two most important columns: Complaint type and Address. Then, we'll create a brand new dataframe 
# just for grouping and cumulating all the complaints to each address
bronx_df_address_complaints = df_bronx_heating[["Incident Address", "Complaint Type"]].groupby(["Incident Address"])
# Pointing number of heating complaints to each address
bronx_heating_complaints = bronx_df_address_complaints.count()
bronx_heating_complaints.head()
```

```python
# Let's also just change the column name to something more meaningful
bronx_heating_complaints.columns = ["NumComplaints"]
bronx_heating_complaints.head(10)
```

### Merge the two dataframes

```python
merged_bx_df = pd.merge(bronx_heating_complaints, bronx_df_, left_index=True, right_on="Address", how="right")
merged_bx_df.head(10) 
```

```python
# Adding a column to indicate if there was a heating complaint.
merged_bx_df["isHeating"] = merged_bx_df["NumComplaints"]
# This column will be identified as a simple true or false
merged_bx_df["isHeating"].replace(np.nan, 0.0, inplace = True)
merged_bx_df["isHeating"] = np.where(merged_bx_df["isHeating"] >  0, True, False)
```

```python
# We'll use to fix all NaN appearing in a few different columns. In the NumComplaints, replace NaN with 0. 
# In the coordinate columns, drop the row entirely, because it's useless to the study
merged_bx_df["NumComplaints"].replace(np.nan, 0.0, inplace = True)

merged_bx_df.dropna(subset = ["YCoord"], axis = 0, inplace=True)
merged_bx_df.dropna(subset = ["ZipCode"], axis = 0, inplace=True)
merged_bx_df.head(10)
```

```python
# Just check for any more null values in the columns
merged_bx_df.isnull().sum()
```

```python
# Now to detect if we have any correlations between values
# Can use Pearson | Kendall | Spearman. Pearson is best for such a wide yet ordinal dataset
merged_bx_df.corr(method="pearson")
```

This initial observation does not suggest a great correlation between any variables. At best, there is a small association between the <u>number of floors</u> in a building and the number of heating complaints.

One last effort would be to see if there's a greater relation only in the cases with more than one heating complaint.

```python
#let's see if anything changes when we drop addresses with 0 complaints.
merged_HeatTrue_bx_df = merged_bx_df[merged_bx_df["NumComplaints"] != 0]
merged_HeatTrue_bx_df.head(10)
```

```python
# Finding correlation between all variables
merged_HeatTrue_bx_df.corr(method="pearson")
```

```python
# Plotting the direct correlation between number of complaints and the variable with the greatest association might provide some insight
plt.scatter(merged_HeatTrue_bx_df["NumFloors"],merged_HeatTrue_bx_df["NumComplaints"],label="Luck")
plt.xlabel('Number of Floors') # add to x-label to the plot
plt.ylabel('Number of Heating Complaints') # add y-label to the plot
plt.title('Number of Heating Complaints corresponding to Building Floor Count from 2010 to 2020') # add title to the plot
plt.show()
```

The plot details two outliers, but even without those there is no obvious connection between the number of floors and the number of heating complaints.

## Conclusion: Based on available data, there is no obvious relationship between building characteristics and the top identified complaint based on evaluations of pearson's correlation. 

### i) There is a small association between the number of floors in a building and the number of heating complaints.
### ii) Other building characteristics with very small associations include: building area, building depth, residential floor area, built floor area ratio, and maximum allowable floor area ratio.
